```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
list_images_names = os.listdir()
```

```python
default_h = 204
for image_name in list_images_names:
    if image_name.endswith('.jpg'):
        
        image = cv2.imread(image_name)
        w2h_scale = image.shape[1]/image.shape[0]
        image = cv2.resize(image,(int(w2h_scale*default_h),default_h))
        cv2.imwrite(image_name, image)
        print(image.shape)
```

```python
default_h = 204
for image_name in list_images_names:
    if image_name.endswith('.jpg'):
        
        image = cv2.imread(image_name)
        if image.shape[1]<204:
            total_w_difference = 204-image.shape[1]
            r_border = np.random.randint(0,total_w_difference)
            l_border = total_w_difference - r_border
            image = cv2.copyMakeBorder(image, 0,0,l_border,r_border,cv2.BORDER_CONSTANT)

            
        elif image.shape[1] > 204:
            total_h_difference = image.shape[1] - 204
            t_border = np.random.randint(0,total_h_difference)
            b_border = total_h_difference - t_border
            image = cv2.copyMakeBorder(image, t_border,b_border,0,0,cv2.BORDER_CONSTANT)
            image = cv2.resize(image, (204,204))
            
        else:
            print(image_name)
            
            
        cv2.imwrite(image_name, image)

            
```
